## Importando las librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Obteniendo datos
En este paso estableceremos un arreglo de columnas para las cabeceras las cuales son:
* BI-RADS
* Age
* Shape
* Margin
* Density
* Severity

Pero la columna **'BI-RADS'** se eliminiará dado que no contribuye a la clasificación

```python
column_names = ['BI-RADS', 'Age', 'Shape', 'Margin', 'Density', 'Severity']
mammographic_masses = pd.read_csv('mammographic_masses.data.txt', names=column_names)
mammographic_masses.drop('BI-RADS',axis=1,inplace=True)
```

### Graficando los datos
Vamos a realizar una serie de gráficas que nos ayuden a entender mejor los datos obtenidos

#### Mapa de calor
Gracias al mapa de calor podremos observar que datos son los faltantes y como están distribuidos en el conjunto de datos.

```python
sns.heatmap(mammographic_masses.isnull(),
            yticklabels = False, cbar = False, cmap = 'viridis')
```

En el gráfico apreciamos que los datos faltantes no siguien un distribución uniforme, por lo que se pueden eliminar sin problema.

#### Gráficos de barra
Nos permitirá observar como los datos se distribuyen de acuerdo a la columna 'Severity'

```python
sns.set_style('whitegrid')
sns.countplot(x='Severity',hue='Shape',data=mammographic_masses,palette='RdBu_r')
```

```python
sns.countplot(x='Severity',hue='Margin',data=mammographic_masses,palette='RdBu_r')
```

```python
sns.countplot(x='Severity',hue='Density',data=mammographic_masses,palette='RdBu_r')
```

### Estandarizar las variables
En este paso se procede a convertir los frames de datos de Pandas en matrices numpy que puede usar scikit_learn.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(mammographic_masses.drop('Severity',axis=1))
```

### Construyendo las matrices
m_caracteristicas es la matriz con los datos de características (edad, forma, margen y densidad) y también se define la matriz_severidad.

```python
scaled_features = scaler.transform(mammographic_masses.drop('Severity',axis=1))
```

```python
m_caracteristicas = pd.DataFrame(scaled_features,columns=mammographic_masses.columns[:-1])
m_caracteristicas.head()
```

```python
matriz_severidad = mammographic_masses['Severity']
```
